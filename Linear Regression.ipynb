```python
import numpy as np
#By default delimiter is white space for loadtxt
data=np.loadtxt("Downloads/data.csv",delimiter=",")
```

```python
#x always needs to be 2d-array
x=data[:,0].reshape(-1,1)
y=data[:,1]
```

```python
x.shape
```

```python
#splitting data
from sklearn import model_selection
X_train,X_test,Y_train,Y_test=model_selection.train_test_split(x,y)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
#Value of m
model.coef_
```

```python
model.intercept_
```

```python
import matplotlib.pyplot as plt
#plotting mx+c line
m=model.coef_[0]
c=model.intercept_
#since value starting from 30
x_line=np.arange(30,70,0.1)
y_line=m*x_line+c
plt.plot(x_line,y_line,"r")
train_1d=X_train.reshape(75)
plt.scatter(X_train,Y_train)
plt.show()
```

```python
#test data
x_line=np.arange(30,70,0.1)
y_line=m*x_line+c
plt.plot(x_line,y_line,"r")
test_1d=X_test.reshape(25)
plt.scatter(X_test,Y_test)
plt.show()
```

We used 75 data points and built a line and the idea is that line should do well for the test data points as well as they are from data set as they follow some patttern


Example-
For data point 35 actual value is around 43 but our model predicted 53. For some data points model is doing well and for some points it might not do well.

```python
score_test=model.score(X_test,Y_test)
```
